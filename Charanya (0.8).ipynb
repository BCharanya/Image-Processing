### Image Processing

```python
import os
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
fort=os.listdir('F:/fort/Letitia wright')
gulf=os.listdir('F:/gulf/ed sheeran')
jeff=os.listdir('F:/jeff/jeff satur')
```

```python

limit = 50
fort_images =[None]*limit
j=0
for i in fort:
    if(j<limit):
        fort_images[j]=imread("F:/fort/Letitia wright/"+i)
        j+=1
    else:
        break
```

```python
limit = 50
gulf_images =[None]*limit
j=0
for i in gulf:
    if(j<limit):
        gulf_images[j]=imread("F:/gulf/ed sheeran/"+i)
        j+=1
    else:
        break
```

```python
limit = 50
jeff_images =[None]*limit
j=0
for i in jeff:
    if(j<limit):
        jeff_images[j]=imread("F:/jeff/jeff satur/"+i)
        j+=1
    else:
        break
```

```python
actor_1=pd.concat([fort_data,gulf_data])
actor=pd.concat([actor_1,jeff_data])
actor
```

```python
from sklearn.utils import shuffle
Thai_indexed=shuffle(actor).reset_index()
Thai_indexed
```

```python
Thai_actors=Thai_indexed.drop(['index'],axis=1)
Thai_actors
x=Thai_actors.values[:,:-1]
y=Thai_actors.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
clf=svm.SVC()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
y_pred
```

```python
for i in (np.random.randint(0,6,4)):
    predicted_images=(np.reshape(x_test[i],(512,512)).astype(np.float64))
    plt.title('predicted label:{0}'.format(y_pred[i]))
    plt.imshow(predicted_images,interpolation='nearest',cmap='gray')
    plt.show()
```

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
